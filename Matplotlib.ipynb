# Libraries

```python
import matplotlib.pyplot as plt
from random import randint as rnd
```

```python
x = list(range(50))
y = [rnd(-10,10) for i in range (50)]
z = [rnd(-10,10) for i in range (50)]



plt.figure(figsize = (20,5)) #should be first command aslo 20 and 5 are best numbers for jupyternotebook
plt.title("Random numbers")
plt.xlabel("time")
plt.ylabel("value")
plt.plot(x,y, label = "Gas price", color = "blue") #also it is possible to use rgb code for colors search for thatذ
plt.plot(x,z, label = "Gas price", color = "Orange")

plt.legend()
plt.grid()
plt.show()
```

# Bar Plot

```python
plt.bar(x,y)
plt.show()
#in bar plot x axes can take non-numercial values
```

```python
plt.barh(x,y)
plt.show()
#in barh plot y axes can take non-numercial values
```

# Pie Plot

```python
names = ["taha", "ara", "pouria"]
vals = [15,130,50]
colors = ["red" ,"green", "purple"]
plt.pie(vals, labels = names, colors = colors, startangle=90, autopct="%2.2f%%" ,hatch=['/', 'o', '--'])
#plt.savefig() always before plt.show(). otherwise only a blank image will be saved
plt.show()
```

# Scatter plot

```python
def r_points(number, _min, _max):
    points = [rnd(_min, _max) for i in range(number)]
    return points
```

```python
b_age = r_points(13,0,18)
b_speed = r_points(13,0,115)


o_age = r_points(15,0,18)
o_speed = r_points(15,0,115)

plt.scatter(b_age, b_speed ,label = "bmw", alpha = 0.7)
plt.scatter(o_age, o_speed, label = "RMA", alpha = 0.7)

plt.legend()
plt.ylabel("speed")
plt.xlabel("age")
plt.title("Speed age")

plt.grid()
plt.show()
```
